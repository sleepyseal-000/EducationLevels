```python
data = [['Bronx',20.3],['Brooklyn',38.8], ['Manhattan',62.2], ['Queens',33.5], ['Staten Island',20.0]] 
df = pd.DataFrame(data, columns = ['name', 'percent'])
```

```python
nyc_boroughts['features'][0].keys()
```

```python
nyc_boroughts['features'][1]['properties']
```

```python
bo_id_map = {}
for feature in nyc_boroughts['features']:
    feature['id'] = feature['properties']['boro_code']
    bo_id_map[feature['properties']['boro_name']] = feature['id']
```

```python
bo_id_map
```

```python
df['id']= df['name'].apply(lambda x: bo_id_map [x])
```

```python
final_data = px.choropleth_mapbox(df, geojson=nyc_boroughts, locations='id', color='percent',  
color_continuous_scale="Viridis",range_color=(0, 80), mapbox_style="carto-positron", center = {"lat": 40.74, "lon": -73.9888}, zoom=9, opacity=0.9, labels={'percent':'Percentage of 25 years and over that has a bachelors degree or higher'})

final_data.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
final_data.show()
```
