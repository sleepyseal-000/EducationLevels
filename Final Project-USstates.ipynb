```python
import json
import pandas as pd
import plotly.express as px
```

```python
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
usa_states = json.load(open("states.geojson","r"))
```

```python
df= pd.read_csv("educated_states.csv")
```

```python
df
```

```python
usa_states['features'][0].keys()
```

```python
state_id_map = {}
for feature in usa_states['features']:
    feature['id'] = feature['properties']['STATEFP']
    state_id_map[feature['properties']['NAME']] = feature['id']
```

```python
state_id_map
```

```python
df['id']= df['State'].apply(lambda x: state_id_map[x])
```

```python
df
```

```python
final= px.choropleth(df,locations="id", geojson=usa_states, color="educationScore", scope="usa", color_continuous_scale="Viridis",)
final.show()
```
